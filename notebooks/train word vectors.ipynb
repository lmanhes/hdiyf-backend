```python
import sys, os
sys.path.append("..")
import gensim
from gensim.models import Word2Vec
import spacy
import pandas as pd
import pickle

from scripts.train_w2v import Corpus
```

```python
core_model = spacy.load('en_core_web_sm', exclude=['ner'])
```

```python
dataset = pd.read_csv("../data/clean_fake_news_dataset.csv")
```

```python
corpus = Corpus(core_model=core_model, dataset=dataset)
```

```python
sentences = corpus()
```

```python
with open("../data/w2v_sentences.pkl", "wb") as f:
    pickle.dump(sentences, f)
```

```python
vocabulary = corpus.get_vocabulary()
```

```python
with open("../data/w2v_vocabulary.pkl", "wb") as f:
    pickle.dump(vocabulary, f)
```

```python
model = Word2Vec(sentences,
                 min_count=5,   # Ignore words that appear less than this
                 size=100,      # Dimensionality of word embeddings
                 window=5,      # Context window for words during training
                 iter = 3)
```

```python
model.wv.most_similar('politic')
```

```python
model.wv.save_word2vec_format('../models_files/w2v_model.bin', binary=True)
```
